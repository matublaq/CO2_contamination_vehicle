```python
# Config. b치sica de Chrome. (Lo abre en modo headless, es decir sin interfaz gr치fica).
options = webdriver.ChromeOptions()
options.add_argument('--headless=') # Opcional, no abre la ventana del navegador.
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=options)
try: 
    # Cargo la p치gina web en Selenium BeautifulSoup.
    driver.get("https://www.eea.europa.eu/en/datahub/datahubitem-view/fa8b1229-3db6-495d-b18e-9c9b3267c02b")
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Cuantos anos de datos hay disponibles.
    years_data = soup.find_all('div', class_="ui pointing secondary vertical menu")[0] # Hay 2 con el mimso class. Nos interesa el 0. 
    years_data = [x.getText() for x in years_data.find_all('a')] #Obtenemos tdos los anos disponibles. 

    ############################
    # Entramos al link donde se encuentran los datos a descargar. 
    driver.get("http://co2cars.apps.eea.europa.eu")
    time.sleep(3)

    # Desplegar el seleccionable para seleccionar todos. (Son los years disponibles). 
    desplegable_years = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="co2_year"]/span')) # "/html/body/div[3]/div/div[1]/div[2]/div[1]/div/div/div/div[2]/div[1]/div[1]/div[2]/h2/span"
    )
    desplegable_years.click()

    # Esperar a que se carguen todos los years en el desplegable.
    WebDriverWait(driver, 3).until(
        EC.presence_of_all_elements_located((By.XPATH, '//*[@id="j1_1"]')) # "/html/body/div[3]/div/div[1]/div[2]/div[1]/div/div/div/div[2]/div[1]/div[1]/div[2]/div/ul/li"
    )
    time.sleep(1)
    # Seleccionar todos los years disponibles.
    list_years = driver.find_elements((By.XPATH, '//*[@id="j1_1"]'))


finally: 
    driver.quit()

driver.quit()
```
